```python
from torchvision import datasets, transforms
from torch.utils.data import DataLoader, random_split

# Define transforms
transform = transforms.Compose([
    transforms.Resize((64, 64)),
    transforms.ToTensor(),
    transforms.Normalize([0.5]*3, [0.5]*3)
])

# Load full dataset
dataset = datasets.ImageFolder("/content/dataset", transform=transform)

# ✅ Split into 80% training and 20% testing
train_size = int(0.8 * len(dataset))
test_size = len(dataset) - train_size
train_dataset, test_dataset = random_split(dataset, [train_size, test_size])

# Create DataLoaders
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False)

# Check number of classes and total images
print("Classes:", dataset.classes)
print("Total images:", len(dataset))
print("Training images:", len(train_dataset))
print("Testing images:", len(test_dataset))
```

CNN Structure

```python
class MyBollywoodCNN(nn.Module):
    def __init__(self, num_classes):
        super(MyBollywoodCNN, self).__init__()

        self.conv_layers = nn.Sequential(
            nn.Conv2d(3, 32, kernel_size=3, padding=1),
            nn.ReLU(),  # ✅ FIXED
            nn.BatchNorm2d(32),
            nn.MaxPool2d(kernel_size=2, stride=2),

            nn.Conv2d(32, 64, kernel_size=3, padding=1),
            nn.ReLU(),  # ✅ FIXED
            nn.BatchNorm2d(64),
            nn.MaxPool2d(kernel_size=2, stride=2),

            nn.Conv2d(64, 128, kernel_size=3, padding=1),
            nn.ReLU(),  # ✅ FIXED
            nn.BatchNorm2d(128),
            nn.MaxPool2d(kernel_size=2, stride=2),
        )

        self.fullyconnected_layers = nn.Sequential(
            nn.Flatten(),
            nn.Linear(128 * 8 * 8, 512),
            nn.ReLU(),  # ✅ FIXED
            nn.Linear(512, num_classes)
        )

    def forward(self, x):
        x = self.conv_layers(x)
        x = self.fullyconnected_layers(x)
        return x
```

Create your model and Automatically use GPU if available

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

num_classes = len(dataset.classes)  # or manually set like num_classes = 100

model = MyBollywoodCNN(num_classes=num_classes).to(device)
```

Learning rate and epoch

```python
epochs=20
learning_rate=0.001
```

```python
model = MyBollywoodCNN(num_classes=len(dataset.classes)).to(device)
```

Loss Function

```python
criterion = nn.CrossEntropyLoss()
```

Optimizer

```python
optimizer= torch.optim.Adam(model.parameters(), lr=learning_rate)
```

TRAINING LOOP

```python
# training loop ───────────────────────────────────────────
model.train()                                     # set training mode

for epoch in range(epochs):                       # ➊ outer loop → epochs
    total_epoch_loss = 0.0

    for batch_features, batch_labels in train_loader:   # ➋ inner loop → batches
        batch_features = batch_features.to(device)      # move to GPU/CPU
        batch_labels   = batch_labels.to(device)

        # forward pass
        outputs = model(batch_features)

        # calculate loss
        loss = criterion(outputs, batch_labels)

        # back‑propagation
        optimizer.zero_grad()   # reset gradients
        loss.backward()         # compute gradients

        # update weights
        optimizer.step()

        total_epoch_loss += loss.item()

    avg_loss = total_epoch_loss / len(train_loader)
    print(f"Epoch {epoch+1}/{epochs} | Loss: {avg_loss:.4f}")
```

Evaluation Code

```python
correct = 0
total = 0

with torch.no_grad():  # no backprop
    for images, labels in test_loader:
        images, labels = images.to(device), labels.to(device)
        outputs = model(images)
        _, predicted = torch.max(outputs.data, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

print(f"✅ Test Accuracy: {100 * correct / total:.2f}%")
```

Gradio Code

```python
import gradio as gr, torch, os, glob, random, numpy as np
from PIL import Image

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device).eval()

preprocess   = transform                 # ← training-time transform
class_names  = dataset.classes
dataset_root = "/content/dataset"        # adjust if your celeb folders live elsewhere
def list_images(folder, exts=(".jpg", ".jpeg", ".png", ".bmp", ".webp")):
    # search every sub-folder too
    return [
        p for p in glob.glob(os.path.join(folder, "**", "*"), recursive=True)
        if p.lower().endswith(exts)
    ]


def identify_celebrity(user_img: Image.Image):
    """
    Returns:
      (output_img, message)
        - output_img: numpy image array  OR  None
        - message   : "" on success, or explanation on failure
    """
    try:
        user_img = user_img.convert("RGB")
        x = preprocess(user_img).unsqueeze(0).to(device)

        with torch.no_grad():
            pred_id = model(x).argmax(1).item()

        celeb_name = class_names[pred_id]
        celeb_dir  = os.path.join(dataset_root, celeb_name)
        pics       = list_images(celeb_dir)

        if not pics:
            return None, f"Predicted **{celeb_name}**, but no reference images found."

        ref_path = random.choice(pics)
        print("✅ returning:", ref_path)

        ref_img = Image.open(ref_path).convert("RGB")
        return np.array(ref_img), ""                              # success 👍

    except Exception as e:
        import traceback; traceback.print_exc()
        return None, f"⚠️ {type(e).__name__}: {e}"                # graceful error

# ── Gradio interface: 1st output = image, 2nd = info text
demo = gr.Interface(
    fn        = identify_celebrity,
    inputs    = gr.Image(type="pil", label="Upload your face photo"),
    outputs   = [
        gr.Image(type="numpy", label="Closest celebrity image"),
        gr.Text(label="Info / Warnings")
    ],
    title       = "Bollywood Celebrity Look-alike",
    description = "Upload a face photo; the CNN predicts which Bollywood celebrity you resemble."
)

demo.launch(share=True, debug=True)
```
